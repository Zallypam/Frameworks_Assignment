```python
# -----------------------------------------------
# CORD-19 Metadata Analysis (Parts 1 - 3)
# -----------------------------------------------

# Step 1: Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import seaborn as sns
import os # Import os for path joining

# (Optional: Makes plots look nicer)
plt.style.use('ggplot')

# Define the path to the metadata CSV in the sample data folder
metadata_csv_path = '/content/sample_data/metadata.csv'

# -----------------------------------------------
# Part 1: Data Loading and Basic Exploration
# -----------------------------------------------

# Load the metadata CSV
try:
    df = pd.read_csv(metadata_csv_path, low_memory=False)

    # View first few rows
    print("First 5 rows:")
    display(df.head()) # Use display for better formatting

    # DataFrame dimensions
    print("\nDataset Dimensions (rows, columns):", df.shape)

    # Column data types
    print("\nColumn Data Types:")
    print(df.dtypes)

    # Check for missing data
    print("\nMissing Values Per Column:")
    print(df.isnull().sum())

    # Basic statistics (numerical columns)
    print("\nBasic Statistical Summary:")
    display(df.describe()) # Use display for better formatting

except FileNotFoundError:
    print(f"Error: metadata.csv not found at {metadata_csv_path}. Please ensure the file exists in the specified path.")
except Exception as e:
    print(f"An error occurred during data loading: {e}")


# -----------------------------------------------
# Part 2: Data Cleaning and Preparation
# -----------------------------------------------

# Check if df was loaded successfully before proceeding
if 'df' in locals() and df is not None:
    try:
        # Convert publish_time column to datetime
        df['publish_time'] = pd.to_datetime(df['publish_time'], errors='coerce')

        # Extract Year
        df['year'] = df['publish_time'].dt.year

        # Create new column: abstract word count
        df['abstract_word_count'] = df['abstract'].astype(str).apply(lambda x: len(x.split()))

        # Remove rows missing title or publish_time (important for analysis)
        df_clean = df.dropna(subset=['title', 'publish_time']).copy() # Use .copy() to avoid SettingWithCopyWarning

        print("\nCleaned Dataset Shape:", df_clean.shape)

    except Exception as e:
        print(f"An error occurred during data cleaning: {e}")

    # -----------------------------------------------
    # Part 3: Analysis and Visualization
    # -----------------------------------------------

    # Check if df_clean was created successfully before proceeding
    if 'df_clean' in locals() and df_clean is not None and not df_clean.empty:
        try:
            # 1. Publications by Year
            year_counts = df_clean['year'].value_counts().sort_index()

            plt.figure(figsize=(10,5))
            plt.bar(year_counts.index, year_counts.values)
            plt.title("Publications by Year")
            plt.xlabel("Year")
            plt.ylabel("Number of Papers")
            plt.show()

            # 2. Top Publishing Journals
            top_journals = df_clean['journal'].value_counts().head(10)

            plt.figure(figsize=(10,5))
            top_journals.plot(kind='bar')
            plt.title("Top 10 Journals Publishing COVID-19 Research")
            plt.xlabel("Journal")
            plt.ylabel("Paper Count")
            plt.xticks(rotation=45, ha='right')
            plt.show()

            # 3. Word Cloud of Titles
            titles_text = " ".join(df_clean['title'].astype(str))
            wordcloud = WordCloud(width=1200, height=600, background_color='white').generate(titles_text)

            plt.figure(figsize=(12, 6))
            plt.imshow(wordcloud, interpolation='bilinear')
            plt.axis('off')
            plt.title("Word Cloud of Paper Titles")
            plt.show()

            # 4. Publications by Source
            source_counts = df_clean['source_x'].value_counts().head(10)

            plt.figure(figsize=(10,5))
            source_counts.plot(kind='bar')
            plt.title("Top Data Sources in CORD-19 Metadata")
            plt.xlabel("Source")
            plt.ylabel("Number of Papers")
            plt.xticks(rotation=45, ha='right')
            plt.show()

        except Exception as e:
            print(f"An error occurred during analysis and visualization: {e}")
    else:
        print("Data cleaning failed or resulted in an empty DataFrame. Skipping analysis and visualization.")
else:
    print("Data loading failed. Skipping data cleaning, analysis, and visualization.")
```
